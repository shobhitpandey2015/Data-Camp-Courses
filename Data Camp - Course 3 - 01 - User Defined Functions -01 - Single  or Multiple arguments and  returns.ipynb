```python
#Defining out own function is often useful. Let's define a function that squares a number. Syntax has two parts: function header and function body

def square():  # <-- Function header
    new_value = 4 ** 2    # <-- Function body
    print(new_value) 

#right now the square() function does not have any parameter, we will add them later. Right now it will always return 4**2 whih is 16

square()  # <-- calling the function
```

```python
 #2) Define functions with single parameters

#What if we wanted to square any number we like. TO add this functinoality, we need to add parameter while defining function and later pass an argument while calling it.

## --- When we define functions, we right paramters in function header
## --- When we call a function, we pass an argument into the function

def square1(value):  # <-- Function header, fucntion which takes one parameter
    new_value = value ** 2    # <-- Function body
    print(new_value) 
    
square1(17) 


#Above function takes one argument, and print the value. What if we dont want to print the value, rather retunr the value.
```

```python
#3) Define funtions that return a value

# To return a value via a single paramter user defined function

def square2(value):  # <-- Function header, fucntion which takes one parameter
    new_value = value ** 2    # <-- Function body
    return new_value          # <-- Function returns the value which could be assigned instead of printing it.

y = square2(5)
print(y)
```

```python
## -- DOCSTRINGS :
        # describes what your funnction does
        # these serve as documentation for your function
        # Function docstrings are placed in immediate line after the function header in """ """
        
def square2(value):  # <-- Function header, fucntion which takes one parameter
    """Return a square value"""
    new_value = value ** 2    # <-- Function body
    return new_value          # <-- Function returns the value which could be assigned instead of printing it.

z = square2(9)
print(z)
```

```python
#important to see differene in paramter and argument writing. Example of one paramter which require string argument while calling

# Define shout with the parameter, word
def shout(word):
    """Print a string with three exclamation marks"""
    # Concatenate the strings: shout_word
    shout_word = word + '!!!'

    # Print shout_word
    print(shout_word)

# Call shout with the string 'congratulations'
shout("congratulations")  # <-- Note just shout(congratulations) does not work
```

```python
#4) Define function with multiple arguments

#Let's define a function that raises the value of a value to another value. Two arguments need to be passed, so two parameters in definition

def power(value1, value2): # <-- Function header line, two parameters defined
    """Raises value1 to the power of value2"""
    output = value1** value2
    return output #<-- retunrs one output

k = power(3,4) # <-- calling by passing 2 arguments, the order is important
print(k)
```

```python
#4) Define function with multiple return values -- TUPLES 

    # A function can provide multiple retunr values by using TUPLES
    # TUPLES -- like list, can contains multiple values. 
    #        -- however unlike list it is IMMUTABLE (cant modify values)
    #        -- constructed using () instead of []  used for lists
    #        -- tuples can be unpacked into several variables in one line. Doing so we assign the variables the tuples value in order of definintion
    #        -- Tuples can be access/ sliced just like list tuple[index] for calling specific tuple element with the index 
    
tuple1 = (2,3,4) # <-- Defining TUPLES and assigning its values
print(tuple1)
type(tuple1)
```

```python
a,b,c = tuple1 # <-- UNPACKING TUPLES 

print(a)
print(b)
print(c)
```

```python
print(tuple1[1]) #Accesing tuple to print the second element
```

```python
#Using TUPLES to construt multiple FUNCTIONS RETURNING MULTIPLE VALUES

# Example function which returns power of one value to the other and vice versa.

def raise_both(value1, value2): # <-- Function header line, two parameters defined
    """Raises value1 to the power of value2 and vice versa"""  #<-- Docstring
    output1 = value1** value2  # <-- Function body
    output2 = value2** value1
    new_tuple = (value1, value2)  # <-- defining tuple with both output
    return new_tuple #<-- retunrs one output #<-- returning multiple values using tuple

answer = raise_both(2,3) #<-- return a tuple and store as a tuple variable
print(answer)

answer1 = raise_both(2,3)[0]  #<-- return a tuple and store as individual elements by slicing the returned tuple
answer2 = raise_both(2,3)[1]
print(answer1)
print(answer2)
```

```python
## Lets put it all together to write a function that analyzes twitter data. 

#This function will return a dictionary, containing data of how many times s a certain language was used across all the tweets in the dataframe
# The function would return Languages as keys and number of times it was used as values
# We will later (next file) learn how to generalize such a function for count the occurances of any item in a DataFrame network
# The dataset contains Twitter data and you will iterate over entries in a column to build a dictionary...
# ... in which the keys are the names of languages and the values are the number of tweets in the given language. 

#We will import the twitter DataFrame file using pandas.
#Doing this without a user-defined function

import pandas as pd

df = pd.read_csv("D:\Data Camp\Course 3-01-Tweets.csv")  # <-- # Import Twitter data as DataFrame: df
print(df)
```

```python
langs_count = {} # <-- Initialize an empty dictionary: langs_count


col = df['lang'] # Extract language column from DataFrame: col

# Iterate over lang column in DataFrame
for entry in col:

    if entry in langs_count.keys():    # If the language is in langs_count, add 1 
        langs_count[entry] = langs_count[entry] + 1
    else:                              # Else add the language to langs_count, set the value to 1
        langs_count[entry] = 1

# Print the populated dictionary
print(langs_count)
```

```python
## Performing above task using a user defined function which Return a dictionary with counts of occurrences as value for each key

# Define count_entries()
def count_entries(df, col_name):  # <-- Two parameters, dataframe and string type parameters
    """Return a dictionary with counts of 
    occurrences as value for each key."""

    # Initialize an empty dictionary: langs_count
    langs_count = {}
    
    # Extract column (of languages keys) from DataFrame: col
    col = df[col_name]  #Note how string automatically is passed here to clice dicionary
    
    # Iterate over lang column in DataFrame
    for entry in col:

        # If the language is in langs_count, add 1
        if entry in langs_count.keys():
            langs_count[entry] =  langs_count[entry] + 1
        # Else add the language to langs_count, set the value to 1
        else:
            langs_count[entry] = 1

    # Return the langs_count dictionary
    return langs_count

# Call count_entries(): result
result = count_entries(df, "lang" )

# Print the result
print(result)
```
