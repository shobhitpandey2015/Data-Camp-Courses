```python
# 1) Plain *txt files

# This could be plain text files like a novel page 
# something containing table data  -- this is know as FLAT FILES


        ## 1.1) READING A PLAIN TEXT FILE - using open() function -- require close() function at the end
    
filename = "random_text.txt"  #<-- needs to be current working directory

file = open(filename, mode = 'r') # r is read only mode

text = file.read() #<-- applying the method read() to read the entire text file

file.close() #<-- close the connection to the file

print(text)

print(file.closed) #<-- # Check whether file is closed
```

```python
        ## 1.2) READING A PLAIN TEXT FILE - using context manager with and open() function - dosen't require close()
    
# Context manager with allows you to create a context in which we can execute commands with the file open

filename = "random_text.txt"

with open(filename, mode = 'r') as file:
    print(file.read())

# <-- file closed automatically outside the context of the with clause
# <-- What we are doing here is binding a variable in the context manager construct 
# .. while still within this construct the variable file will be bound to open filenmae in read mode
```

```python
    #2) FLAT FILES 
    
        #2.1) INTRODUCTION TO FLAT FILES
         
            # Flat files are basic text files containing records, that is table data without structured relationships
         # This is in contrast to relational databases, in which coloumns of different tables can be related
         # More generically flat files consist of : 
            # Records : A record in a flat file is composed of fields or attributes, each of which contains at most one item of information. Ex : name of persons
            # Coloumn : contians feature or attributes. Ex: age, complexity, height of the persons
            # Header : could be present. It's the firt row, describing the contents of the data coloumns or states of what the corresponding coloumns and there featured attrubtes are. Ex. Age, weight, etc coloumn names
         # File Extension :  
            # *csv - Comma Separated Values -- values in each row (records) are separated in comma 
            # *txt -- text file
         # Delimiter : Values in flat files can be separated by various characters called delimiters. ex comma (','), tabs ('\t')
        
        
        ## How do you import flat files?
            # Two important packages - NumPy, Pandas
            # Numpy - If a file consists entirely of numbers and we want to store as Numpy arrays
            # Pandas - when we want to store data into a dataframe
```

```python
      # 2.2) Importing Flat files with only numeric data (or simple mixed datatype) using Numpy
    
# Example MNIST.txt which is comma delimited and importing it as an Numpy Array
# Why Numpy Array : 
    # Numpy arrays are python's standard for storing numerical data
    # Numpy arrays are also essential for other packages such as Sci-Kit learn (used for machine learning)
    # Consists of various Numpy functions to load data efficiently into Numpy arryas. Example :
        # np.loadtxt() -
            # -- great for simple importing where data type is not mixed
            # -- arguments : filename, delimiter, skiprows, usecols, dtype and many other arguments
            
        # np.genfromtxt() -
            # -- can handle mixed data types. 
            # -- If we pass dtype=None to it, it will figure out what types each column should be.
            # -- results in a structured array :  1D array, where each element of the array is a row of the flat file imported
        
        # np.recfromcsv()
            # -- same as np.genfromtxt() except its default dtype = None
        
        
#importing file as numpy array using loadtxt()

import numpy as np

filename = 'data2.csv'
 
data = np.loadtxt(filename, delimiter = ',', skiprows = 1)  #< -- specify the delimter type. By defaults it is blank space
                                                            #< -- skiprows = 1; skips the first row which is string data
print(data)
```

```python
#to extract just the 3 and 4th coloumn
data = np.loadtxt(filename, delimiter = ',', skiprows = 1, usecols = [2,3]) #<-- extract only 3rd and 4th coloumns

#< -- usecols = [] passes the list containing the ints for the index of  coloumns to be extracted 
print(data)
```

```python
# we can also the specify the data type in which we want to import the data

data = np.loadtxt(filename, delimiter = ',', skiprows = 1, usecols = [2,3], dtype = str) 
#<-- extract 3rd and 4th coloumns as strings

#< -- usecols = [] passes the list containing the ints for the index of  coloumns to be extracted 
print(data)
```

```python
# to handle mixed dataypes we used np.genfrmtxt()

data = np.genfromtxt('titanic_sub.csv', delimiter=',', names=True, dtype=None)

# Here, the first argument is the filename, 
# the second specifies the delimiter , 
# names = True -- tells us there is a header. 
# Because the data are of different types, data is an object called a structured array. 
# Because numpy arrays have to contain elements that are all the same type, the structured array solves this by being a 1D array,
#.. where each element of the array is a row of the flat file imported. You can test this by checking out the array's shape in 
#.. the shell by executing np.shape(data)


print(np.shape(data)) #<-- checks the the structured array data type
print(type(data)) # <-- type of output from such mixed type files via genfromtxt()
print(data)

```

```python
# Accessing rows and columns of structured arrays is super-intuitive: 
# to get the ith row, merely execute data[i] and to get the column with name 'Fare', execute data['Fare']

print(data['Survived']) #<-- only getting the survived coloumn from the structured numpy array data 
```

```python
# to handle mixed dataypes we can also use  np.recfromcsv()

# Assign the filename: file
file = 'titanic_sub.csv'

# Import file using np.recfromcsv: d
d = np.recfromcsv(file)

# Print out first three entries of d
print(d[:3])
```
